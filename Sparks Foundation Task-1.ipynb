Preparing the data

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Checking the data 

```python
from sklearn.linear_model import LinearRegression
mo = LinearRegression()
mo.fit(X,y)
```

```python
print(mo.intercept_)
print(mo.coef_)
```

Plotting the Regression Line

```python
line =( mo.coef_*X+mo.intercept_ )
plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

Making Predictions

```python
print(X_test)
y_pred = mo.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

Making Predictions For the Student who Studies for 9.25 hrs

```python
pred_score = mo.predict([[9.25]])
print("The prediction score is :",pred_score)
```

Evaluating the Model

```python
from sklearn import metrics
print('Mean Absolute Erorr:',metrics.mean_absolute_error(y_test, y_pred))
```
